# What would be good noise distributions for AM, FM, PM noise examples?

```python
a0 = 1.
f0 = 9.192168670e9
p0 = 0
y1 = a0 * np.sin( 2 * np.pi * f0 * t + p0)
plt.plot(t, y1)
plt.title('Original signal')
```

```python
a0 = 1. + normal_noise
f0 = 9.192168670e9
p0 = 0
y1 = a0 * np.sin( 2 * np.pi * f0 * t + p0)
plt.plot(t, y1)
plt.title('Amplitude noise example')
```

```python
a0 = 1.
f0 = 9.192168670e9 * (1 + 0.2 * normal_noise)
p0 = 0
y1 = a0 * np.sin( 2 * np.pi * f0 * t + p0)
plt.plot(t, y1)
plt.title('Frequency noise example')
```

```python
a0 = 1.
f0 = 9.192168670e9
p0 = 0
y1 = a0 * np.sin( 2 * np.pi * f0 * t + p0 + 5 * normal_noise)
plt.plot(t, y1)
plt.title('Phase noise example')
```
